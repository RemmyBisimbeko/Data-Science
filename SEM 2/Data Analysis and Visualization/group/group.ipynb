## Method One

### Defining the Variables

```python
# Defining the variables with several data points
x1 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
x2 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
x3 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
x4 = np.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8])

y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
y2 = np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
y3 = np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])
y4 = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])

#the above arrays from x1 to y4 are organized into a dictionary (datasets)
datasets = {'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4, 'y1': y1, 'y2': y2, 'y3': y3, 'y4': y4}

# Convert the above dictionary into a DataFrame (df)
df = pd.DataFrame(datasets)
```

### Calculating Descriptive Statistics

```python

descriptive_stats = df.describe()
```

```python
print("Descriptive Statistics:\n\n", descriptive_stats)
```

### Calculating correlations

```python
correlations = df.corr()
```

```python
print("\nCorrelations:\n", correlations)
```

### Calculate linear regression parameters for each dataset

```python
regression_results = {}  #Initialize Dictionary for Regression Results

#Perform Linear Regression for Each Dataset
for i in range(1, 5):
    slope, intercept, r_value, p_value, std_err = stats.linregress(df[f'x{i}'], df[f'y{i}'])
    
    regression_results[f'regression_{i}'] = {
        'slope': slope,
        'intercept': intercept,
        'r_value': r_value,
        'p_value': p_value,
        'std_err': std_err
    }
```

slope:  The slope of the regression line. <br>
intercept:  The y-intercept of the regression line.<br>
r_value:  Correlation coefficient.<br>
p_value:  p-value for a hypothesis test whose null hypothesis is that the slope is zero.<br>
std_err:  Standard error of the estimated gradient.<br>

### Displaying Regression Results

```python
print("\nRegression Results:\n", regression_results)
```

### Plotting scatter plots with regression lines

```python
# Plotting scatter plots with regression lines
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

for i, ax in enumerate(axs.flat, 1):
    sns.scatterplot(x=df[f'x{i}'], y=df[f'y{i}'], ax=ax)

    slope = regression_results[f'regression_{i}']['slope']
    intercept = regression_results[f'regression_{i}']['intercept']

    x = np.linspace(df[f'x{i}'].min(), df[f'x{i}'].max(), 100) 
    y = slope * x + intercept

    ax.plot(x, y, color='red')
    ax.set_title(f'Dataset {i}: y = {slope:.2f}x + {intercept:.2f}')
    ax.set_xlabel(f'x{i}')
    ax.set_ylabel(f'y{i}')

plt.tight_layout()
plt.show()
```

## Method Two

### Read the Data using Pandas

```python
dataframe = pd.read_csv("anscombe dataset.csv")
```

### Calculating Descriptive Statistics

```python
descriptive_stats = dataframe.describe()
descriptive_stats
```

### Calculating the Correlations Between the Different Variables

```python
correlations = dataframe.corr()
correlations
```

### Calculate linear regression parameters for each dataset

```python
regression_results = {} #Initialize Dictionary for Regression Results

#Perform Linear Regression for Each Dataset
for i in range(1, 5):
    slope, intercept, r_value, p_value, std_err = stats.linregress(dataframe[f'x{i}'], dataframe[f'y{i}'])
    
    regression_results[f'regression_{i}'] = {
        'slope': slope,
        'intercept': intercept,
        'r_value': r_value,
        'p_value': p_value,
        'std_err': std_err
    }
```

### Displaying Regression Results

```python
print("\nRegression Results:\n", regression_results)
```

### Plotting scatter plots with regression lines

```python
# Plotting scatter plots with regression lines
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

for i, ax in enumerate(axs.flat, 1):
    sns.scatterplot(x=dataframe[f'x{i}'], y=dataframe[f'y{i}'], ax=ax)

    slope = regression_results[f'regression_{i}']['slope']

    intercept = regression_results[f'regression_{i}']['intercept']

    x = np.linspace(dataframe[f'x{i}'].min(), dataframe[f'x{i}'].max(), 100)
    y = slope * x + intercept

    ax.plot(x, y, color='red')
    ax.set_title(f'Dataset {i}: y = {slope:.2f}x + {intercept:.2f}')
    ax.set_xlabel(f'x{i}')
    ax.set_ylabel(f'y{i}')

plt.tight_layout()
plt.show()
```
