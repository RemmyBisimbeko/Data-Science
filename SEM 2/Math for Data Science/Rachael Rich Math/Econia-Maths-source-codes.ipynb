1. **(i) Calculate the sample correlation coefficient and comment on its value.**

   The sample correlation coefficient (Pearson's correlation coefficient) $ r $ is given by:
  $$
   r = \frac{n \sum xy - (\sum x)(\sum y)}{\sqrt{\left[n \sum x^2 - (\sum x)^2\right] \left[n \sum y^2 - (\sum y)^2\right]}}
  $$

   Substituting the given values:
   $$
   r = \frac{9 \cdot 2341.6 - (422 \cdot 48.7)}{\sqrt{[9 \cdot 20132 - (422)^2][9 \cdot 276.91 - (48.7)^2]}}
   $$

   Simplifying further:
   $$
   r = \frac{21074.4 - 20551.4}{\sqrt{[181188 - 178084][2492.19 - 2371.69]}}
   $$
   $$
   r = \frac{523}{\sqrt{3104 \cdot 120.5}}
   $$
   $$
   r = \frac{523}{\sqrt{374032}}
   $$
   $$
   r = \frac{523}{373.8}
   $$
   $$
   r \approx 0.8552
   $$

   Comment: The value of $ r $ indicates a strong positive correlation between mobility rate and divorce rate. Since the value is close to 1. This suggests that there a stronger linear correlation between the two variables.

2. **(ii) Calculate the regression equation.**

   The regression equation of $ y $ on $ x $ is given by:
   $$
   y = a + bx
   $$
   where:
   $$
   b = \frac{n \sum xy - (\sum x)(\sum y)}{n \sum x^2 - (\sum x)^2}
   $$
   $$
   a = \frac{\sum y - b \sum x}{n}
   $$

   Using the given values:
   $$
   b = \frac{9 \cdot 2341.6 - 422 \cdot 48.7}{9 \cdot 20132 - 422^2}

   $$

   $$

   $$
   a = \frac{48.7 - 0.1685 \cdot 422}{9}
   $$
   $$
   a = \frac{48.7 - 71.107}{9}
   $$
   $$
   a \approx -2.49
   $$

   Thus, the regression equation is:
   $$
   y = -2.49 + 0.1685x

**(b) Write down the condition for three events $ A, B $, and $ C $ to be independent.**

Three events $ A, B $, and $ C $ are independent if:
   $$
   P(A \cap B \cap C) = P(A) \cdot P(B) \cdot P(C)
   $$
   Additionally, the pairwise independence conditions must also hold:
   $$
   P(A \cap B) = P(A) \cdot P(B)
   $$
   $$
   P(B \cap C) = P(B) \cdot P(C)
   $$
   $$
   P(A \cap C) = P(A) \cdot P(C)
   $$

**Are the following statements "True" or "False"?**

$$ P(A) = 0.4 $$

$$ P(A|C) = \frac{P(C|A) \cdot P(A)}{P(C)} = \frac{0.25 \cdot 0.4}{0.1} = 1 $$

Since $ P(A|C) = 1 $, it means that $ C $ implies $ A $.

**Answer:** True

To check if $ B $ and $ C $ are mutually exclusive, we need to verify if $ P(B \cap C) = 0 $. This can be inferred from $ P(B|C) = 0 $, which implies:

$$ P(B|C) = \frac{P(B \cap C)}{P(C)} = 0 $$
$$ \implies P(B \cap C) = 0 $$

**Answer:** True

#### (b) What is $ P(B|A) $?

$$ P(B|A) = \frac{P(A \cap B)}{P(A)} $$

We can use the given $ P(A|B) $ and $ P(B) $:

$$ P(A|B) = \frac{P(A \cap B)}{P(B)} $$

$$ P(A \cap B) = P(A|B) \cdot P(B) = 0.2 \cdot 0.5 = 0.1 $$

#### (c) What is $ P(B|A^c) $?

#### (d) Two fair dice are thrown.

**(i) Determine the probability distribution of the sum of the two dice, $ X $, and find its mean and variance.**

The possible sums $ X $ of two dice range from 2 to 12. The probability distribution can be derived from the number of ways to achieve each sum:

$$
\begin{aligned}
&X = 2: \quad 1 \text{ way } (1,1) \\
&X = 3: \quad 2 \text{ ways } (1,2), (2,1) \\
&X = 4: \quad 3 \text{ ways } (1,3), (2,2), (3,1) \\
&X = 5: \quad 4 \text{ ways } (1,4), (2,3), (3,2), (4,1) \\
&X = 6: \quad 5 \text{ ways } (1,5), (2,4), (3,3), (4,2), (5,1) \\
&X = 7: \quad 6 \text{ ways } (1,6), (2,5), (3,4), (4,3), (5,2), (6,1) \\
&X = 8: \quad 5 \text{ ways } (2,6), (3,5), (4,4), (5,3), (6,2) \\
&X = 9: \quad 4 \text{ ways } (3,6), (4,5), (5,4), (6,3) \\
&X = 10: \quad 3 \text{ ways } (4,6), (5,5), (6,4) \\
&X = 11: \quad 2 \text{ ways } (5,6), (6,5) \\
&X = 12: \quad 1 \text{ way } (6,6) \\
\end{aligned}
$$

The probability of each sum $ X = k $ is $ P(X = k) = \frac{\text{Number of ways to get } k}{36} $.

$$
\begin{aligned}
&P(X = 2) = \frac{1}{36} \\
&P(X = 3) = \frac{2}{36} = \frac{1}{18} \\
&P(X = 4) = \frac{3}{36} = \frac{1}{12} \\
&P(X = 5) = \frac{4}{36} = \frac{1}{9} \\
&P(X = 6) = \frac{5}{36} \\
&P(X = 7) = \frac{6}{36} = \frac{1}{6} \\
&P(X = 8) = \frac{5}{36} \\
&P(X = 9) = \frac{4}{36} = \frac{1}{9} \\
&P(X = 10) = \frac{3}{36} = \frac{1}{12} \\
&P(X = 11) = \frac{2}{36} = \frac{1}{18} \\
&P(X = 12) = \frac{1}{36} \\
\end{aligned}
$$

**Mean ($ \mu $) and Variance ($ \sigma^2 $)**

$$ \mu = \frac{1}{36} \cdot 252 = 7 $$

Variance:
$$ \sigma^2 = \sum_{k=2}^{12} (k - \mu)^2 \cdot P(X = k) $$

$$ \sigma^2 = \sum_{k=2}^{12} (k - 7)^2 \cdot P(X = k) $$

$$ \sigma^2 = (2 - 7)^2 \cdot \frac{1}{36} + (3 - 7)^2 \cdot \frac{2}{36} + \ldots + (12 - 7)^2 \cdot \frac{1}{36} $$

$$ \sigma^2 = \frac{1}{36} [(5)^2 + 2 \cdot (4)^2 + 3 \cdot (3)^2 + 4 \cdot (2)^2 + 5 \cdot (
 1)^2 + 6 \cdot (0)^2 + 5 \cdot (1)^2 + 4 \cdot (2)^2 + 3 \cdot (3)^2 + 2 \cdot (4)^2 + (5)^2] $$

$$ \sigma^2 = \frac{1}{36} [25 + 32 + 27 + 16 + 5 + 0 + 5 + 16 + 27 + 32 + 25] $$
