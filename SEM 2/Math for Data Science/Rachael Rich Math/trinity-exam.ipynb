### (c) Given a random variable $X$ with PDF $ f(x) = \lambda e^{-\lambda x} $ for $ x \geq 0 $, derive the expressions for mean and variance of $X$.

For an exponential distribution with PDF $ f(x) = \lambda e^{-\lambda x} $:

- **Mean ( $ \mu $ ):** 
  $$
  E(X) = \int_0^\infty x \lambda e^{-\lambda x} \, dx = \frac{1}{\lambda}
  $$

**Detailed derivation (maybe unnecessary)**

Certainly! Let's delve deeper into the calculation of the expected value (mean) of an exponential random variable.


An exponential random variable $X$ with parameter $\lambda$ has the probability density function (PDF):

$$ f(x) = \lambda e^{-\lambda x}, \quad x \geq 0 $$

The expected value (mean) of $X$ is given by:

$$ E(X) = \int_0^\infty x \cdot f(x) \, dx $$

Substitute the PDF $f(x)$ into the integral:

$$ E(X) = \int_0^\infty x \cdot \lambda e^{-\lambda x} \, dx $$

Using integration by parts to solve this integral. --Integration by parts is given by:

$$ \int u \, dv = uv - \int v \, du $$

Let:

$$ u = x \quad \Rightarrow \quad du = dx $$
$$ dv = \lambda e^{-\lambda x} \, dx \quad \Rightarrow \quad v = -e^{-\lambda x} $$

Using these, the integration by parts formula becomes:

$$ \int x \lambda e^{-\lambda x} \, dx = x (-e^{-\lambda x}) \Big|_0^\infty - \int (-e^{-\lambda x}) \, dx $$

Evaluating the first term:

$$ x (-e^{-\lambda x}) \Big|_0^\infty = \left[ -x e^{-\lambda x} \right]_0^\infty $$

As $ x \to \infty $, $ x e^{-\lambda x} $ approaches 0 because the exponential term $ e^{-\lambda x} $ decays much faster than the linear term $ x $ grows. So,

$$ \left[ -x e^{-\lambda x} \right]_0^\infty = 0 - (0 \cdot e^{0}) = 0 - 0 = 0 $$

Evaluating the integral $ \int -e^{-\lambda x} \, dx $:

$$ \int -e^{-\lambda x} \, dx = \frac{-e^{-\lambda x}}{-\lambda} = \frac{e^{-\lambda x}}{\lambda} $$

Suubstituting this back into our integration by parts formula:

$$ \int_0^\infty x \lambda e^{-\lambda x} \, dx = 0 - \left[ \frac{e^{-\lambda x}}{\lambda} \right]_0^\infty $$

Evaluate this term:

$$ \left[ \frac{e^{-\lambda x}}{\lambda} \right]_0^\infty = \frac{e^{-\lambda \cdot \infty}}{\lambda} - \frac{e^{-\lambda \cdot 0}}{\lambda} = \frac{0}{\lambda} - \frac{1}{\lambda} = 0 - \frac{1}{\lambda} = -\frac{1}{\lambda} $$

Thus, we have:

$$ E(X) = \int_0^\infty x \lambda e^{-\lambda x} \, dx = \frac{1}{\lambda} $$

Therefore, the expected value (mean) of an exponential random variable $X$ with parameter $\lambda$ is $ \frac{1}{\lambda} $.

A large manufacturing company has recently signed a deal to manufacture trail mix for a well-known food label. This label makes 3 versions of its product - one for airlines, one for grocery stores, and one for gas stations. Each version has a different mixture of peanuts, raisins, and chocolate which serves as the base of the trail mix. The base mixtures are made in 15 kg batches and sent to a second building for packaging. The following table contains the information about the mixes, each row containing the recipe for a 15 kg batch.

|            | Raisins (kg/batch) | Peanuts (kg/batch) | Chocolate (kg/batch) | Sale Price ($/kg) |

Given that the total mixtures of Raisins, Peanuts, and Chocolate are 380, 5500, and 620 kgs, respectively:

(ii) If the manufacturer wanted to use up all the ingredients in storage each day, how many batches of each mixture (airline, gas station, and grocery) should be made?

The total mixtures of Raisins, Peanuts, and Chocolate can be expressed as the following system of linear equations:

1. Raisins: $7x + 2y + 6z = 380$

#### LU Decomposition of a Matrix $ A $

**LU decomposition** is a factorization of a matrix $ A $ into the product of a lower triangular matrix $ L $ and an upper triangular matrix $ U $. This decomposition can be written as:

$$ A = LU $$

where:
- $ L $ is a lower triangular matrix (all elements above the diagonal are zero).
- $ U $ is an upper triangular matrix (all elements below the diagonal are zero).

#### Significance of LU Decomposition in Data Science

1. **Solving Systems of Linear Equations**: LU decomposition simplifies the process of solving linear equations. Given $ Ax = b $, we can write $ LUx = b $. By defining $ y = Ux $, we solve $ Ly = b $ and then $ Ux = y $. This is computationally efficient.

2. **Matrix Inversion**: LU decomposition can be used to find the inverse of a matrix. Once $ A $ is decomposed into $ L $ and $ U $, the inverse can be found by inverting $ L $ and $ U $ and then multiplying them.

3. **Determinant Calculation**: The determinant of $ A $ can be easily computed from $ L $ and $ U $. Since $ A = LU $, $ \text{det}(A) = \text{det}(L) \cdot \text{det}(U) $. For triangular matrices, the determinant is the product of the diagonal elements, making it straightforward.

4. **Numerical Stability**: LU decomposition helps in improving the numerical stability of algorithms. Decomposing $ A $ into $ L $ and $ U $ reduces rounding errors in numerical computations.

5. **Eigenvalues and Eigenvectors**: LU decomposition is a step in algorithms for finding eigenvalues and eigenvectors, which are important in many data science applications such as PCA and spectral clustering.

6. **Optimization Problems**: Many optimization algorithms, such as those used in machine learning and statistical inference, require the solution of linear systems or matrix inversions, where LU decomposition is a critical tool.

**Solution:**

To find the probability that the sample mean among 1000 customers differs from the true mean by more than 5 minutes, we'll use the Central Limit Theorem (CLT). Here are the steps to solve this:

Given:
- Population standard deviation ($ \sigma $) of monthly minutes = 120 minutes.
- Sample size ($ n $) = 1000.

### Step 1: Calculate the Standard Error of the Mean
The standard error of the mean ($ SE $) is given by:
$$ SE = \frac{\sigma}{\sqrt{n}} $$

Substitute $ \sigma = 120 $ minutes and $ n = 1000 $:
$$ SE = \frac{120}{\sqrt{1000}} $$
$$ SE = \frac{120}{31.62} $$
$$ SE \approx 3.79 $$

### Step 2: Determine the Probability using the Normal Distribution
The probability that the sample mean differs from the true mean by more than 5 minutes can be approximated using the normal distribution, due to the CLT.

Let $ \bar{X} $ be the sample mean. We are interested in finding $ P(|\bar{X} - \mu| > 5) $.

$$ P(|\bar{X} - \mu| > 5) = P\left(\left|\frac{\bar{X} - \mu}{SE}\right| > \frac{5}{SE}\right) $$
$$ = P\left(\left|\frac{\bar{X} - \mu}{3.79}\right| > \frac{5}{3.79}\right) $$
$$ \approx P\left(\left|\frac{\bar{X} - \mu}{3.79}\right| > 1.32\right) $$

### Step 3: Calculate the Z-score and Find the Probability




```python


# Given data
sigma = 120  # Standard deviation of monthly minutes
n = 1000     # Sample size
SE = sigma / (n ** 0.5)  # Standard error of the mean

# Margin of error in minutes
margin_of_error = 5

# Calculate the Z-score
z_score = margin_of_error / SE

# Calculate the probability using the complementary CDF
probability = 1 - stats.norm.cdf(z_score)

print(f"Z-score: {z_score:.4f}")
print(f"Probability that |X - | > 5 minutes: {probability:.4f}")
```

- $ \hat{p} $ is the estimated proportion.
