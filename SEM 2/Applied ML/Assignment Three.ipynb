```python
# Extract only the fields/features we want in our training Data Set
# training_dataset = pd.DataFrame(final_df).drop(columns='origin')

feature_dataset = final_df[['mpg','cylinders','displacement','horsepower','weight','acceleration']]
print(feature_dataset)
```

```python
# Now sub divide our data set into training and test Data Sets
# 0.8 = 80%, good practice to always have the training data higher than the test data
training_dataset = feature_dataset.sample(frac=0.8, random_state=0) 
testing_dataset = feature_dataset.drop(training_dataset.index)
```

```python
# Replace '?' with NaN
training_dataset.replace('?', np.nan, inplace=True)

# Convert columns to numeric, forcing errors to NaN
training_dataset[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'mpg']] = training_dataset[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'mpg']].apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values
training_dataset.dropna(inplace=True)
```
