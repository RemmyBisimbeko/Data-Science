### 3.3.4. Feature selection</div>
Feature selection is a crucial step in building a machine learning model, and the choice of the best method often depends on the specific characteristics of the data at hand. Here are some common feature selection methods we considered for our credit card fraud detection.

#### 1. Recursive Feature Elimination (RFE)
RFE is a backward selection approach where the model is trained on all features, and the least important features are recursively eliminated until the desired number of features is reached, and Logistic Regression is often used as the base model for RFE.Amongst its strengths it captures interactions between features, it an iterative process as it can provide insight into feature importance.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Recursive Feature Elimination (RFE)
model_rfe = LogisticRegression()
rfe_selector = RFE(model_rfe, n_features_to_select = 20)
X_train_rfe = rfe_selector.fit_transform(X_train, y_train)
X_test_rfe = rfe_selector.transform(X_test)

# Printing headers of selected features
selected_features_rfe = X_train.columns[rfe_selector.support_]
print("Selected Features (RFE):")
print(selected_features_rfe)
```

#### 2. Feature Importance from Tree-based Model (Random Forest)
Random Forest calculates feature importance based on the decrease in impurity (Gini index) when a particular feature is used for splitting, Random Forest Classifier is used to train the model. The strength of the method is its efficient for both classification and regression tasks,and captures non-linear relationships.

## 5.1. Hyperparameter Tuning Logistic Regression Model with Base Model</div>

i. C which is the (Inverse of regularization strength)
   - Searching for the optimal value for C. Important to note here higher values can lead to overfitting, while lower values can result in underfitting.

ii. Penalty
   - Choosing between L1 and L2 regularization. Depending on the nature of our data and whether feature selection is important.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_curve, auc

# Creating a logistic regression model
lrg = LogisticRegression()

# Defining the hyperparameters to search
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
}

# Creating GridSearchCV
grid_search = GridSearchCV(lrg, param_grid, cv = 5, scoring = 'f1', verbose = 1)

# Fiting the model to the resampled data
grid_search.fit(X_train, y_train)

# Printing the best parameters and corresponding F1 score
print("Best Parameters: ", grid_search.best_params_)
print("Best Cross-validated F1 Score: {:.2f}".format(grid_search.best_score_))

# Evaluating the model on the original test set
y_pred_lrg_grid_srch = grid_search.best_estimator_.predict(X_test)

# Printing classification report and confusion matrix
print("\nClassification Report:\n", classification_report(y_test, y_pred_lrg_grid_srch))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_lrg_grid_srch))

# Calculating Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, grid_search.best_estimator_.predict_proba(X_test)[:, 1])
prc_auc = auc(recall, precision)

# Calculate ROC Curve
fpr, tpr, _ = roc_curve(y_test, grid_search.best_estimator_.predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

# Plot Precision-Recall Curve and ROC Curve on the same row
plt.figure(figsize = (16, 6))

# Subplot for Precision-Recall Curve
plt.subplot(1, 2, 1)
plt.plot(recall, precision, color = 'blue', label = f'PR Curve (AUC = {prc_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc = 'lower right')

# Subplot for ROC Curve
plt.subplot(1, 2, 2)
plt.plot(fpr, tpr, color = 'red', label = f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='black', linestyle = '--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc = 'lower right')
```

The ROC Curve with an Area Under the Curve (AUC) of 0.89 suggests that the model has a strong ability to discriminate between the positive and negative classes. The higher AUC value indicates good performance in terms of the true positive rate and false positive rate trade-off.

On the other hand, the PR Curve with an AUC of 0.56 reflects a moderate precision-recall trade-off. This means that while the model performs reasonably well, there is room for improvement, especially in optimizing precision and recall simultaneously.

iii. Min Samples Leaf
   - Defining the minimum number of samples required to be at a leaf node.

```python
# Creating a decision tree model
dctr = DecisionTreeClassifier(random_state = 42)

# Defining the hyperparameters to search
param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Creating GridSearchCV
grid_search_dctr = GridSearchCV(dctr, param_grid, cv = 5, scoring = 'f1', verbose = 1)

# Fitting the model to the data
grid_search_dctr.fit(X_train, y_train)

# Printing the best parameters and corresponding F1 score
print("Best Parameters: ", grid_search_dctr.best_params_)
print("Best Cross-validated F1 Score: {:.2f}".format(grid_search_dctr.best_score_))

# Evaluating the model on the test set
y_pred_dctr = grid_search_dctr.best_estimator_.predict(X_test)

# Printing classification report and confusion matrix
print("\nClassification Report:\n", classification_report(y_test, y_pred_dctr))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_dctr))

# Calculating Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, grid_search_dctr.best_estimator_.predict_proba(X_test)[:, 1])
prc_auc = auc(recall, precision)

# Calculating ROC Curve
fpr, tpr, _ = roc_curve(y_test, grid_search_dctr.best_estimator_.predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

# Plot Precision-Recall Curve and ROC Curve on the same row
plt.figure(figsize = (16, 6))

# Subplot for Precision-Recall Curve
plt.subplot(1, 2, 1)
plt.plot(recall, precision, color = 'blue', label = f'PR Curve (AUC = {prc_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc = 'lower right')

# Subplot for ROC Curve
plt.subplot(1, 2, 2)
plt.plot(fpr, tpr, color = 'red', label = f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='black', linestyle = '--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc = 'lower right')
```

The ROC Curve's AUC of 0.86 indicates that the model has a good ability to discriminate between the positive and negative classes. The higher AUC value suggests a favorable balance between the true positive rate and the false positive rate.

Meanwhile, the PR Curve's AUC of 0.63 implies a moderate precision-recall trade-off. The model performs reasonably well, but there is room for improvement, particularly in optimizing precision and recall simultaneously.
