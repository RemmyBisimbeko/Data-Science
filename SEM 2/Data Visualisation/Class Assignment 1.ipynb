```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Define Columns: Data
x1 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
x2 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
x3 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
x4 = np.array([8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8])

y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
y2 = np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
y3 = np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73])
y4 = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89])
```

```python
# Specify the dataset
datasets = {'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4, 'y1': y1, 'y2': y2, 'y3': y3, 'y4': y4}
```

```python

# Convert to DataFrame
df = pd.DataFrame(datasets)
```

```python
# Calculate Descriptive Statistics
descriptive_stats = df.describe()
```

```python
# Calculate correlations
correlations = df.corr()
```

```python
# Calculate linear regression parameters for each dataset
regression_results = {}
for i in range(1, 5):
    slope, intercept, r_value, p_value, std_err = stats.linregress(df[f'x{i}'], df[f'y{i}'])
    regression_results[f'regression_{i}'] = {
        'slope': slope,
        'intercept': intercept,
        'r_value': r_value,
        'p_value': p_value,
        'std_err': std_err
    }
```

```python
# Plotting scatter plots with regression lines
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

for i, ax in enumerate(axs.flat, 1):
    sns.scatterplot(x=df[f'x{i}'], y=df[f'y{i}'], ax=ax)
    slope = regression_results[f'regression_{i}']['slope']
    intercept = regression_results[f'regression_{i}']['intercept']
    x = np.linspace(df[f'x{i}'].min(), df[f'x{i}'].max(), 100)
    y = slope * x + intercept
    ax.plot(x, y, color='red')
    ax.set_title(f'Dataset {i}: y = {slope:.2f}x + {intercept:.2f}')
    ax.set_xlabel(f'x{i}')
    ax.set_ylabel(f'y{i}')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Display descriptive statistics and correlations
print("Descriptive Statistics:\n", descriptive_stats)
print("\nCorrelations:\n", correlations)
print("\nRegression Results:\n", regression_results)
```
