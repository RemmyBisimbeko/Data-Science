```python
# Import necessary libraries
import requests  # For making HTTP requests
from bs4 import BeautifulSoup  # For parsing HTML content
import pandas as pd  # For data manipulation and storage
```

```python
# URL of the webpage to scrape
url = 'https://unche.or.ug/all-academic-programs/'
```

```python

# Function to scrape the academic programs table
def scrape_academic_programs(url):
    """
    This function scrapes the academic programs table from the given URL
    and returns the data as a pandas DataFrame.
    """
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    # Redundant
    if response.status_code != 200: 
        raise Exception(f"Failed to load page {url}")
    
    # print(response)
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table on the webpage
    table = soup.find('table')
    
    # Extract the headers
    headers = [header.get_text(strip=True) for header in table.find_all('th')]
    
    # Extract the rows
    rows = []
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cells = row.find_all('td')
        rows.append([cell.get_text(strip=True) for cell in cells])
    
    # Create a DataFrame from the extracted data
    df = pd.DataFrame(rows, columns=headers)
    
    return df
```

```python
# Scrape the academic programs table
academic_programs_df = scrape_academic_programs(url)

# Save the DataFrame to a CSV file
csv_file_path = 'scrapper.csv'
academic_programs_df.to_csv(csv_file_path, index=False)

# Display the DataFrame
print("Scraped DataFrame:")
print(academic_programs_df)

# Read the contents of the CSV file
read_df = pd.read_csv(csv_file_path)

# Display the contents read from the CSV file
print("\nContents of scrapper.csv:")
read_df.head(10)
```
