#### 8. Lyrics Finder
* Lyrics Finder using a lyrics API: Create a lyrics finder tool that retrieves song lyrics from a lyrics API based on user input, allows users to search for lyrics by song title, artist, or album.


Let's start implementing by selecting a lyrics API and then setting up the Python environment. 
For this project, we'll use the Genius API, which provides access to a large database of song lyrics.

```python
# Import necessary libraries
import requests
import pandas as pd
import seaborn as sns
import lyricsgenius
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

This code provides a basic framework for the Lyrics Finder tool using the Genius API. It defines functions to search for song lyrics, parse HTML content to extract lyrics, and generate a word cloud visualization of the lyrics. The main() function demonstrates an example usage by searching for lyrics of the song "Sitya ~Loss" by Eddy Kenzo.

```python
# 
#Function to search for song lyrics
def search_song_lyrics(title, artist):
    try:
        # Initialize Genius API client
        genius = lyricsgenius.Genius("X9PNUnAFtBMaR-NiQNZ83sskqLE70Uvn23xGdMmujLCquAiBPVzRjBLDFfVs-wv3")
        
        # Search for the song lyrics
        song = genius.search_song(title, artist)
        if song:
            return song.lyrics
        else:
            print("Song lyrics not found.")
            return None
    except Exception as e:
        print("Error occurred while searching for lyrics:", e)
        return None

# Step 3: Function to generate word cloud from lyrics
def generate_wordcloud(lyrics):
    # Generate word cloud from the lyrics
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(lyrics)

    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

Let's use the song "Sitya Loss" by Eddy Kenzo as an example. 
Finally, we generate a word cloud visualization of the lyrics.

```python
# Step 4: Main function to execute the project
def main():
    # Search for lyrics of the song "Sitya Loss" by Eddy Kenzo
    title = "Sitya Loss"
    artist = "Eddy Kenzo"
    lyrics = search_song_lyrics(title, artist)
    if lyrics:
        print("Lyrics found:")
        print(lyrics)
        # Generate word cloud from the lyrics
        generate_wordcloud(lyrics)
    else:
        print("Failed to fetch song lyrics.")

# Execute the main function
if __name__ == "__main__":
    main()
```
