#### 3. Movie Recommender
* Movie Recommender using movie database API: Build a movie recommendation system that uses a movie database API to suggest movies based on user preferences, ratings, genres, and past viewing history.

We start by importing necessary libraries such as requests, pandas, numpy, matplotlib, seaborn, and warnings.

```python
# Import necessary libraries
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Ignore warnings
warnings.filterwarnings("ignore")
```

The fetch_movie_data function retrieves movie data from a movie database API using the provided API key.
The explore_movie_data function explores the fetched movie data, including previewing the first few rows, describing the data, and checking for null values.
The visualize_movie_data function visualizes the movie data using histograms for rating distribution and bar charts for genre frequency.

```python
# Function to fetch movie data from API
def fetch_movie_data(api_key):
    """
    Fetches movie data from a movie database API.
    
    Parameters:
    - api_key: API key for the movie database API
    
    Returns:
    - DataFrame containing movie data
    """
    # Make API request to fetch movie data
    url = f'https://api.themoviedb.org/3/discover/movie?api_key=de84d59e44e5149c6a859c06cdd0d997&language=en-US&sort_by=popularity.desc&include_adult=false&include_video=false&page=1'
    response = requests.get(url)
    
    # Convert response to DataFrame
    if response.status_code == 200:
        data = response.json()
        movies = pd.DataFrame(data['results'])
        return movies
    else:
        print("Failed to fetch movie data.")
        return None

# Function to explore movie data
def explore_movie_data(movies_df):
    """
    Explores movie data including preview, description, and null values.
    
    Parameters:
    - movies_df: DataFrame containing movie data
    """
    # Preview the first few rows of the DataFrame
    print("Preview of movie data:")
    display(movies_df.head())
    
    # Describe the DataFrame to understand its structure
    print("\nDescription of movie data:")
    display(movies_df.describe())
    
    # Check for null values
    print("\nNull values in movie data:")
    display(movies_df.isnull().sum())

# Function to visualize movie data
def visualize_movie_data(movies_df):
    """
    Visualizes movie data using graphs and charts.
    
    Parameters:
    - movies_df: DataFrame containing movie data
    """
    # Plot rating distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(movies_df['vote_average'], bins=20, kde=True, color='skyblue')
    plt.title('Rating Distribution')
    plt.xlabel('Rating')
    plt.ylabel('Frequency')
    plt.show()

    # Plot genre frequency
    plt.figure(figsize=(12, 8))
    genre_counts = movies_df['genre_ids'].explode().value_counts().sort_values(ascending=False)
    genre_counts.plot(kind='bar', color='lightgreen')
    plt.title('Genre Frequency')
    plt.xlabel('Genre')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

The main function orchestrates the execution of the entire notebook, fetching movie data, exploring it, and visualizing it.
Finally, I execute the main function to run the entire notebook.

```python
# Main function
def main():
    # Set API key
    api_key = 'de84d59e44e5149c6a859c06cdd0d997' 
    
    # Fetch movie data
    movies_df = fetch_movie_data(api_key)
    if movies_df is not None:
        # Explore movie data
        explore_movie_data(movies_df)
        
        # Visualize movie data
        visualize_movie_data(movies_df)

# Execute main function
if __name__ == "__main__":
    main()
```
